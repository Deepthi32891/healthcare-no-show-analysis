```python
# No-Show Appointments Analysis Project

## Objective:
#Analyze patient appointment data to uncover patterns related to no-shows and understand the impact of factors 
#like SMS reminders, age, and day of the week.



## Step 1: Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For cleaner visuals
sns.set(style="whitegrid")



##  Step 2: Load the Data

df = pd.read_csv("No-show appointments.csv")
df.head()



##  Step 3: Data Cleaning

# Convert date columns to datetime
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])

# Standardize column names
df.columns = df.columns.str.strip().str.replace("-", "_").str.lower()

# Drop irrelevant columns
df.drop(['patientid', 'appointmentid'], axis=1, inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Add 'waiting_days' column
df['waiting_days'] = (df['appointmentday'] - df['scheduledday']).dt.days



## Step 4: Exploratory Data Analysis

# 1. No-show distribution
no_show_rate = df['no_show'].value_counts(normalize=True) * 100
print("\nNo-show Rates (%):\n", no_show_rate)

# 2. Impact of SMS reminders
sms_effect = df.groupby('sms_received')['no_show'].value_counts(normalize=True).unstack()
print("\nNo-show by SMS Received:\n", sms_effect)

# 3. Most affected age group
bins = [0, 12, 18, 35, 50, 65, 100]
labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Middle Age', 'Senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# 4. No-shows by age group
age_group_analysis = df[df['no_show'] == 'Yes']['age_group'].value_counts()
print("\nNo-shows by Age Group:\n", age_group_analysis)

# 5. Day of week effect
df['day_of_week'] = df['appointmentday'].dt.day_name()
no_shows_by_day = df[df['no_show'] == 'Yes']['day_of_week'].value_counts()



## Step 5: Visualizations

# No-show rates pie chart
no_show_rate.plot(kind='pie', autopct='%1.1f%%', title='Appointment Show vs No-Show', ylabel='')
plt.show()

# No-show % by weekday
plt.figure(figsize=(8,5))
sns.countplot(data=df[df['no_show'] == 'Yes'], x='day_of_week', order=df['day_of_week'].value_counts().index)
plt.title('No-shows by Day of the Week')
plt.xticks(rotation=45)
plt.ylabel('No-show Count')
plt.xlabel('Day of Week')
plt.show()

```
